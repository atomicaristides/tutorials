```python
import matplotlib.pyplot as plt
import numpy as np

E_f = 3.04*10**(-11) #J/fission Energy per fission
Sigma_f = 0.333 #cm^-1
R_e = 1
d = 9.5/1000 #diameter mm
r = d/2
H_e = 2.69
z = np.linspace(-H_e/2,H_e/2,100)
phi_center =  3.4*10**(13)

phi_r_z = phi_center*np.cos(3.14*z/H_e) #neutrons per second per meter squared. 

Q = Sigma_f*E_f*phi_r_z #Watts per cm^3
fig, ax = plt.subplots()  # Create a figure containing a single axes.
plt.plot(z,phi_r_z) # Plot some data on the axes. 
#plt.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Matplotlib plot.

#plt.plot([1, 2, 3, 4])
plt.ylabel('Flux profile, neutron/(cm^-2)')
plt.xlabel('Height, m')
plt.show()
```

```python
plt.ylabel('Volumetric heat generation, Watts/cm^3 ')
plt.plot(z,Q)
plt.xlabel('Height of the core, m')
plt.show()
```
